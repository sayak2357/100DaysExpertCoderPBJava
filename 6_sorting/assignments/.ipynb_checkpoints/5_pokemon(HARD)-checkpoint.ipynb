```python
Pokemon Battle
Arnab is given 
N
 pokemon, his friend Daipayan is also given 
N
 pokemon. Arnab and Daipayan want to fight against each other 
N
 times. After a battle, the pokemon need to rest and they are not available for the upcoming battles. 
    So Daipayan being confident gives Arnab the order he will send the pokemon's to fight. 
    Each pokemon has power, so the pokemon with greater power wins, if the power is equal then it ends in a draw.
Now Arnab wants your help to know how many matches he can win in the best possible scenario.

Input format
First-line contains an integer 
T
, denoting number of test cases.
For each test case :
The first line contains an integer 
N
.
The second line contains 
N
 space-separated integers denoting the power of Arnab's Pokemons.
The third line contains 
N
 space-separated integers denoting the power of Daipayan's Pokemons in the order.

Output format
For each test case print the required answer.

Constraints:
1
<=
t
<=
10

1
<=
n
<=
10
6

Example
Input
1
5
8 4 2 1 9
1 2 3 4 5

Output
4

Explanation
To counter the pokemon of power 
1
 Arnab will send pokemon of power 
2

To counter the pokemon of power 
5
 Arnab will send pokemon of power 
1
 so he loses

To counter the pokemon of power 
3
 Arnab will send pokemon of power 
4

To counter the pokemon of power 
2
 Arnab will send pokemon of power 
9

To counter the pokemon of power 
4
 Arnab will send pokemon of power 
8

so he can win 
4
 matches. 
```

```python
tc = int(input())
for i in range(tc):
    n = int(input())
    arnab = [int(x) for x in input().split(' ')]
    dp = [int(x) for x in input().split(' ')]
    arnab.sort(reverse=True)
    dp.sort(reverse=True)

    wins = 0
    dpointer = 0
    for j in range(n):
        ar = arnab[j]
        while dpointer<n and ar<=dp[dpointer]:
            dpointer+=1

        if dpointer<n:
            wins+=1
            dpointer+=1
    print(wins)
```
